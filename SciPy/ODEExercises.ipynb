For this question refer to the solutions for the simple pendulum we got using scipy's odeint from the beginning of the ODE Tour.

**(a)** See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
from scipy.integrate import odeint

def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])

t = np.arange(0.,10.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0

th0 = 0
th1 = np.pi/4
th2 = np.pi/2
th3 = 3*np.pi/4
th4 = np.pi

theta_init0 = np.array([th0, omega0]) # initial values
theta_init1 = np.array([th1, omega0])
theta_init2 = np.array([th2, omega0])
theta_init3 = np.array([th3, omega0])
theta_init4 = np.array([th4, omega0])

theta0 = odeint(deriv,theta_init0,t)
theta1 = odeint(deriv,theta_init1,t)
theta2 = odeint(deriv,theta_init2,t)
theta3 = odeint(deriv,theta_init3,t)
theta4 = odeint(deriv,theta_init4,t)

plt.plot(t,theta0[:,0],'r--', label = r'$\theta_0 = 0$')
plt.plot(t, theta1[:,0], 'b--', label = r'$\theta_0 = \pi/4$')
plt.plot(t, theta2[:,0], label = r'$\theta_0 = \pi/2$')
plt.plot(t, theta3[:,0], label = r'$\theta_0 = 3\pi/4$')
plt.plot(t, theta4[:,0], label = r'$\theta_0 = \pi$')
plt.xlabel(r'$t(seconds)$', fontsize = 15)
plt.ylabel(r'$\theta (rad)$', fontsize = 15)
plt.legend(loc='best')
plt.show()
```

**(a)** Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
def derivative(initial, t, Vt):
    g = 9.81
    Vx = initial[0]
    Vy = initial[1]
    dVx = (-g)*(Vx/(Vt**2))*np.sqrt(Vx**2 + Vy**2)
    dVy = (-g)*(1+ (Vy/(Vt**2))*np.sqrt(Vx**2 + Vy**2))
    return np.array([dVx, dVy, Vx, Vy])

def projectile(V0, theta0, x0, y0, Vt):
    t = np.arange(0., 200.01, 0.01)
    Vx0 = V0*np.cos(theta0*np.pi/180)
    Vy0 = V0*np.sin(theta0*np.pi/180)
    initial = np.array([Vx0,Vy0, x0, y0])
    velocity = odeint(derivative, initial, t, args = (Vt,))
    plt.plot(velocity[:,2], velocity[:,3])
    plt.ylim(0,30)
    plt.xlim(0, 300)
    plt.xlabel('x', fontsize = 20)
    plt.ylabel('y', fontsize = 20)
    plt.show()
    
projectile(30., 50, 0., 0., 44)
```

(a) Write a function and interact widget that solves the ODE for the nonlinear pendulum with damping and driving. The parameters, ℓ, q, fd, and Ωd should be passed into the interact as sliders. Use initial values of θ0 =π/2, ω0=0 and a time span of 50 seconds.
Check that the solution looks as it did before for the case: ℓ=1, q=0, fd=0, Ωd=0. Play with the sliders to see how the solution depends on the parameters.

```python
def derivNLpen(theta_In, t, ell, q, f, Od):
    g = 9.81
    theta = theta_In[0]
    omega = theta_In[1]
    dtheta = omega
    domega = (-g/ell)*np.sin(theta) - q*omega - f*np.sin(Od * t)
    return array([dtheta, domega])

def NonLinPendulum(theta0 = np.pi/2., omega0 = 0., ell = 1., q = 0., f = 0., Od = 0.):
    t = np.arange(0., 50.01, 0.01)
    theta_In = [theta0, omega0]
    theta = odeint(derivNLpen, theta_In, t, args = (ell, q, f, Od))
    theta1 = theta[:,0]
    omega1 = theta[:,1]
    plt.plot(t, 9.81 * ell*(1- np.cos(theta1))+ 0.5*(ell**2)*(omega1**2), label = 'Energy')
    plt.plot(t,theta[:,0],label='Theta')
    plt.plot(t, theta[:,1], label = 'Omega')
    plt.xlabel(r'$t (seconds)$',fontsize=20)
    plt.ylabel(r'$\theta (rad)/\omega (rad/s)/Energy(J)$',fontsize=20)
    plt.grid()
    plt.legend()
    plt.show()

v = interact(NonLinPendulum, theta0 = (0., np.pi/2.), omega0 = (0., 10.), ell = (0.1, 10.), q = (0., 1.), f = (0., 2.), Od = (0., 1.))
```

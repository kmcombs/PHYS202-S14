**(a)** See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. What happens to the trajectory when the initial angle is 0 and π ?

```python
from scipy.integrate import odeint

def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])

t = np.arange(0.,10.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0

th0 = 0
th1 = np.pi/4
th2 = np.pi/2
th3 = 3*np.pi/4
th4 = np.pi

theta_init0 = np.array([th0, omega0]) # initial values
theta_init1 = np.array([th1, omega0])
theta_init2 = np.array([th2, omega0])
theta_init3 = np.array([th3, omega0])
theta_init4 = np.array([th4, omega0])

theta0 = odeint(deriv,theta_init0,t)
theta1 = odeint(deriv,theta_init1,t)
theta2 = odeint(deriv,theta_init2,t)
theta3 = odeint(deriv,theta_init3,t)
theta4 = odeint(deriv,theta_init4,t)

plt.plot(t,theta0[:,0],'r--')
plt.plot(t, theta1[:,0], 'b-')
plt.plot(t, theta2[:,0])
plt.plot(t, theta3[:,0])
plt.plot(t, theta4[:,0])
plt.show()
```

dydx=x+2y
Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
xmax = 1.25
dx = 0.25
 
nsteps = int(xmax/dx)
x = np.zeros(nsteps)
y = np.zeros(nsteps)
dydx = np.zeros(nsteps)

for i in range(len(x)-1):
    x[i+1] = x[i] + dx
    dydx[i+1] = x[i] + 2*y[i]
    y[i+1] = y[i] + 0.5*(dydx[i]) * dx
    
print y[-1]
```
